# Question #1: Which variables are the most helpful in understanding if a customer is going to be approved or denied for a loan? Are there certain variables that are not useful to collect?

```python
# import necessary libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from imblearn.over_sampling import SMOTE
```

```python
# create deep copy of dataset
data_copy = data.copy(deep = True)
```

```python
data.columns
```

```python
# feature selection
feature_cols = ['Loan_Amount', 'FICO_score',
       'Monthly_Gross_Income', 'Monthly_Housing_Payment',
       'Ever_Bankrupt_or_Foreclose']
# concatenate dummy variables with num_cols
X = pd.concat([data_copy[feature_cols],dummy_data],axis=1)
data2 = pd.concat([X,data_copy[['Approved','Lender']]],axis=1)
```

```python
X.describe(include='all').T
```

```python
# target variable
y = data_copy.Approved
```

```python
# Split dataset into training and test set and include random state for repeatability
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) # 70% training and 30% test
```

```python
# perform oversampling on target variable due to class imbalance in approved
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
```

```python
# Create a Random Forest Classifier
rf=RandomForestClassifier(n_estimators=100, random_state=1)

# Train the model using the training set
rf.fit(X_train_smote,y_train_smote)

# Predict obs in test set
y_pred=rf.predict(X_test)
```

```python
# asses accuracy of random forest model
print('Accuracy:',metrics.accuracy_score(y_test,y_pred))
```

```python
#Look at feature importance
feature_imp = pd.Series(rf.feature_importances_, index=X.columns).sort_values(ascending=False)
#Display feature importance
feature_imp
```

# Question #3: Are there groups of customers that would be a better fit for a different lender? How much incremental revenue could we have made if we matched certain groups of customers more appropriately? 

```python
# Approval characteristics for each lender
approval_summary = data[data['Approved'] == 1].groupby('Lender').agg({
    'FICO_score': ['mean', 'median'],
    'Loan_Amount': ['mean', 'median'],
    'Monthly_Gross_Income': ['mean', 'median']
})

# Identify Matching Criteria for FICO score, Loan Amount, and Monthly Gross Income
matching_criteria_fico = approval_summary['FICO_score']['median'].to_dict()
matching_criteria_loan_amount = approval_summary['Loan_Amount']['median'].to_dict()
matching_criteria_income = approval_summary['Monthly_Gross_Income']['median'].to_dict()

# Revenue per approval for each lender
revenue_per_approved = data[data['Approved'] == 1].groupby('Lender')['bounty'].sum() / data[data['Approved'] == 1].groupby('Lender')['Approved'].count()

# Estimate potential revenue using FICO score, Loan Amount, and Monthly Gross Income
for lender in revenue_per_approved.index:
    median_fico = matching_criteria_fico[lender]
    median_loan_amount = matching_criteria_loan_amount[lender]
    median_income = matching_criteria_income[lender]

    potential_customers = data[(data['FICO_score'] >= median_fico) &
                               (data['Loan_Amount'] >= median_loan_amount) &
                               (data['Monthly_Gross_Income'] >= median_income) &
                               (data['Lender'] != lender)]
    estimated_revenue = potential_customers.shape[0] * revenue_per_approved[lender]
    print(f"Potential incremental revenue for {lender} with FICO, Loan Amount, and Income criteria: ${estimated_revenue:.2f}")

```
